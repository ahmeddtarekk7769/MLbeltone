```python
# Cell 0: Setup NLTK
import nltk
nltk.download('stopwords', download_dir='/tmp/nltk_data')
nltk.download('wordnet', download_dir='/tmp/nltk_data')
nltk.data.path.append('/tmp/nltk_data')
```

```python
# Cell 1: Imports and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
from imblearn.over_sampling import SMOTE
from tqdm import tqdm

tqdm.pandas()

# Global resources
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()
```

```python
# Cell 2: Load Data
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
```

```python
# Cell 3: Class Distribution Plot
label_counts = df_train['label'].value_counts(normalize=True) * 100
plt.figure(figsize=(6, 4))
sns.barplot(x=label_counts.index, y=label_counts.values, palette='pastel')
plt.title('Label Distribution (%)')
plt.ylabel('Percentage')
plt.xlabel('Label')
plt.ylim(0, 100)
for i, v in enumerate(label_counts.values):
    plt.text(i, v + 1, f"{v:.2f}%", ha='center', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Cell 4: Preprocessing Function
def preprocess_text(text):
    text = str(text).lower()
    text = re.sub(r"\d+", "", text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = text.split()
    tokens = [w for w in tokens if w not in stop_words]
    tokens = [lemmatizer.lemmatize(w) for w in tokens]
    return " ".join(tokens)
```

```python
# Cell 5: Apply Preprocessing
df_train['resume_clean'] = df_train['resume_text'].progress_apply(preprocess_text)
df_train['job_clean'] = df_train['job_description_text'].progress_apply(preprocess_text)
df_test['resume_clean'] = df_test['resume_text'].progress_apply(preprocess_text)
df_test['job_clean'] = df_test['job_description_text'].progress_apply(preprocess_text)
```

```python
# Cell 6: Clean & Check Data
print("Missing values in training data:")
print(df_train.isnull().sum())
print("\nMissing values in test data:")
print(df_test.isnull().sum())
print("\nDuplicate rows in training data:", df_train.duplicated().sum())
df_train.drop_duplicates(inplace=True)
```

```python
# Cell 7: Extra Features
df_train['resume_len'] = df_train['resume_clean'].apply(lambda x: len(x.split()))

# Cosine similarity between resume and job
tfidf_sim = TfidfVectorizer()
resume_vecs = tfidf_sim.fit_transform(df_train['resume_clean'])
job_vecs = tfidf_sim.transform(df_train['job_clean'])
df_train['similarity'] = [cosine_similarity(resume_vecs[i], job_vecs[i])[0][0] for i in range(len(df_train))]

# Jaccard similarity
def jaccard_similarity(a, b):
    set_a = set(a.split())
    set_b = set(b.split())
    return len(set_a & set_b) / len(set_a | set_b)
df_train['jaccard'] = df_train.apply(lambda x: jaccard_similarity(x['resume_clean'], x['job_clean']), axis=1)
```

```python
# Cell 8: Vectorize Text
vectorizer = TfidfVectorizer(max_features=1000)
X_text = vectorizer.fit_transform(df_train['resume_clean'] + " " + df_train['job_clean'])
X_text = X_text.toarray()  # convert to dense for standardization

# Combine with extra features
X_features = df_train[['resume_len', 'similarity', 'jaccard']].values
scaler = StandardScaler()
X_features_scaled = scaler.fit_transform(X_features)

X = np.hstack((X_text, X_features_scaled))
label_enc = LabelEncoder()
y = label_enc.fit_transform(df_train['label'])
```

```python
# Cell 9: Handle Class Imbalance with SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
# Cell 10: Topic Modelling
count_vect = CountVectorizer(max_df=0.9, min_df=10, stop_words='english')
dtm = count_vect.fit_transform(df_train['resume_clean'])
lda = LatentDirichletAllocation(n_components=3, random_state=42)
lda.fit(dtm)
def print_topics(model, count_vectorizer, n_top_words):
    words = count_vectorizer.get_feature_names_out()
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic #{topic_idx}: ", " ".join([words[i] for i in topic.argsort()[:-n_top_words - 1:-1]]))

print_topics(lda, count_vect, 10)
```

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

def topic_modeling_per_label(df, text_column, label_column, n_topics=5, n_words=10):
    labels = df[label_column].unique()
    for label in labels:
        print(f"\n{'='*30}\nTopics for label: {label}\n{'='*30}")
        subset = df[df[label_column] == label][text_column]

        count_vect = CountVectorizer(max_df=0.95, min_df=5, stop_words='english')
        dtm = count_vect.fit_transform(subset)

        if dtm.shape[0] < n_topics:
            print(f"Not enough documents for topic modeling with {n_topics} topics.")
            continue

        lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
        lda.fit(dtm)

        words = count_vect.get_feature_names_out()
        for i, topic in enumerate(lda.components_):
            top_words = [words[i] for i in topic.argsort()[:-n_words - 1:-1]]
            print(f"Topic {i+1}: {' | '.join(top_words)}")

# Call this function on your cleaned resume text
topic_modeling_per_label(df_train, text_column='resume_clean', label_column='label')
```

```python
industry_keywords = {
    'Technology': ['software', 'developer', 'python', 'data', 'cloud', 'AI', 'ML'],
    'Finance': ['bank', 'investment', 'financial', 'capital', 'trading'],
    'Healthcare': ['hospital', 'medical', 'health', 'pharmaceutical'],
    'Education': ['teacher', 'school', 'university', 'curriculum', 'student'],
    'Retail': ['store', 'sales', 'retail', 'customer', 'merchandise']
}

def extract_industry(text):
    text = text.lower()
    for industry, keywords in industry_keywords.items():
        if any(keyword in text for keyword in keywords):
            return industry
    return "Other"

df_train['industry'] = df_train['job_description_text'].apply(extract_industry)
```

```python
# View first few rows with extracted info
df_train[['job_description_text', 'job_role', 'industry']]
```

```python
# Cell 11: Train/Test Split
X_train, X_val, y_train, y_val = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
# Cell 12: Train Model
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)
```

```python
# Cell 13: Evaluation
y_pred = clf.predict(X_val)
print(classification_report(y_val, y_pred, target_names=label_enc.classes_))
sns.heatmap(confusion_matrix(y_val, y_pred), annot=True, fmt='d', xticklabels=label_enc.classes_, yticklabels=label_enc.classes_)
plt.title('Confusion Matrix')
plt.show()
```

```python
# Cell 14: Predict on Test Set
X_test_text = vectorizer.transform(df_test['resume_clean'] + " " + df_test['job_clean']).toarray()

# Extract and standardize extra features
test_features = df_test[['resume_text', 'job_description_text']].copy()
test_features['resume_clean'] = df_test['resume_clean']
test_features['job_clean'] = df_test['job_clean']
test_features['resume_len'] = test_features['resume_clean'].apply(lambda x: len(x.split()))

test_resume_vecs = tfidf_sim.transform(test_features['resume_clean'])
test_job_vecs = tfidf_sim.transform(test_features['job_clean'])
test_features['similarity'] = [cosine_similarity(test_resume_vecs[i], test_job_vecs[i])[0][0] for i in range(len(test_features))]
test_features['jaccard'] = test_features.apply(lambda x: jaccard_similarity(x['resume_clean'], x['job_clean']), axis=1)

X_test_feats = test_features[['resume_len', 'similarity', 'jaccard']].values
X_test_feats_scaled = scaler.transform(X_test_feats)

X_test = np.hstack((X_test_text, X_test_feats_scaled))
test_preds = clf.predict(X_test)
df_test['predicted_label'] = label_enc.inverse_transform(test_preds)

# If true labels available
y_test_true = label_enc.transform(df_test['label'])
y_test_pred = label_enc.transform(df_test['predicted_label'])
print(classification_report(y_test_true, y_test_pred, target_names=label_enc.classes_))
```
